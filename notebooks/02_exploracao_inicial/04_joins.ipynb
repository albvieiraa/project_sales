# INNER JOIN

## Pedido + Cliente

```sql
SELECT
  o.order_id,
  o.order_status,
  c.customer_state
FROM workspace.projeto_vendas_bronze.orders o
INNER JOIN workspace.projeto_vendas_bronze.customers c
  ON o.customer_id = c.customer_id
```

```sql
SELECT
  oi.order_id,
  oi.product_id,
  oi.price
FROM workspace.projeto_vendas_bronze.order_items oi
INNER JOIN workspace.projeto_vendas_bronze.orders o
  ON oi.order_id = o.order_id
```

## LEFT JOIN

Clientes cadastrados que nunca fizeram pedido?

```sql
SELECT
  c.customer_id,
  o.order_id
FROM workspace.projeto_vendas_bronze.customers c
LEFT JOIN workspace.projeto_vendas_bronze.orders o
  ON c.customer_id = o.customer_id
```

## JOIN + agregação

### Faturamento por estado

```sql
SELECT
  c.customer_state,
  SUM(oi.price) AS faturamento
FROM workspace.projeto_vendas_bronze.order_items oi
JOIN workspace.projeto_vendas_bronze.orders o
  ON o.order_id = o.order_id
JOIN workspace.projeto_vendas_bronze.customers c
  ON o.customer_id = c.customer_id
GROUP BY c.customer_state
ORDER BY faturamento DESC;
```

### Faturamento por categoria

```sql
SELECT
  p.product_category_name,
  SUM(oi.price) AS faturamento
FROM workspace.projeto_vendas_bronze.order_items oi
JOIN workspace.projeto_vendas_bronze.products p
  ON oi.product_id = p.product_id
GROUP BY p.product_category_name
ORDER BY faturamento DESC;
```
