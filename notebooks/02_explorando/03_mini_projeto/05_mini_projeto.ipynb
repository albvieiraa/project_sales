## Problema de Negócio

Qual o perfil das vendas da Olist, considerando clientes, produtos e faturamento?

### Métricas Gerais

```sql
-- Total de Pedidos
SELECT COUNT(order_id) AS total_pedidos
FROM workspace.projeto_vendas_bronze.orders
```

```sql
-- Total de clientes
SELECT COUNT(DISTINCT customer_unique_id) AS total_clientes
FROM workspace.projeto_vendas_bronze.customers
```

```sql
-- Faturamento total
SELECT SUM (price) AS faturamento_total
FROM workspace.projeto_vendas_bronze.order_items
```

```sql
-- Faturamento de pedidos entregues
SELECT
  SUM(oi.price) AS faturamento_total
FROM
  workspace.projeto_vendas_bronze.order_items AS oi
JOIN workspace.projeto_vendas_bronze.orders AS o ON oi.order_id = o.order_id
WHERE o.order_status = 'delivered';
```

```sql
-- Ticket médio (faturamento total dividido pelo número de pedidos)
SELECT
  SUM(price) / COUNT(DISTINCT order_id) AS ticket_medio
FROM
  workspace.projeto_vendas_bronze.order_items;
```

### Análise Geográfica

```sql
-- Faturamento por estado
SELECT
  c.customer_state,
  SUM(oi.price) AS faturamento
FROM
  workspace.projeto_vendas_bronze.orders AS o
JOIN workspace.projeto_vendas_bronze.customers AS c ON o.customer_id = c.customer_id
JOIN workspace.projeto_vendas_bronze.order_items AS oi ON o.order_id = oi.order_id
GROUP BY
  c.customer_state
ORDER BY
  faturamento DESC
```

### Produtos

```sql
-- Faturamento por categoria de produto
SELECT
  p.product_category_name,
  SUM(oi.price) AS faturamento
FROM
  workspace.projeto_vendas_bronze.order_items AS oi
JOIN workspace.projeto_vendas_bronze.products AS p ON oi.product_id = p.product_id
GROUP BY
  p.product_category_name
ORDER BY
  faturamento DESC
```

## Pedidos não entregues estão impactando o faturamento?

```sql
SELECT
  o.order_status,
  SUM(oi.price) AS faturamento
FROM workspace.projeto_vendas_bronze.order_items oi
JOIN workspace.projeto_vendas_bronze.orders o
  ON oi.order_id = o.order_id
GROUP BY o.order_status
ORDER BY faturamento DESC;
```

A maior parte do faturamento está concentrada em pedidos com status delivered. Pedidos não entregues representam uma parcela significativamente menor do faturamento, indicando que o impacto financeiro direto é limitado. No entanto, esses pedidos podem gerar custos operacionais e impactar a experiência do cliente.

## Existe concentração de vendas em poucas categorias?

O faturamento da Olist apresenta alta concentração em poucas categorias de produtos. As categorias líderes representam uma parcela significativa da receita total, enquanto a maioria das categorias possui participação marginal no faturamento

## Recomendação ao time de negócio

Investir em estratégias para diversificar o faturamento, incentivando categorias com baixo desempenho por meio de campanhas, bundles ou destaque na plataforma.

Monitorar e reduzir pedidos não entregues, não apenas pelo faturamento direto, mas pelo impacto em custos logísticos e na satisfação do cliente.

Priorizar estoque, marketing e parcerias nas categorias líderes, uma vez que elas representam a maior parcela da receita atual.
